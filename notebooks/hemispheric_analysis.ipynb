## Solar Hemispheric Analysis

This notebook's main purpose is to analyze the number of sunspots registered in both hemispheres from 1992 to the present days. The CSV file was obtained from the Solar Influences Data Analysis Center (SIDC). The project functions, in some form, as a sequel to the last repository uploaded in this account regarding the historical sunspot analysis!

## Code

```python
import csv
import matplotlib.pyplot as plt

years = []
north = []
south = []


with open("../data/SN_d_hem_V2.0.csv") as file:
    reader = csv.reader(file, delimiter=';', skipinitialspace=True)
    
    for line in reader:
        years.append(float(line[3]))
        north.append(int(line[5]))
        south.append(-int(line[6]))
```

The code is simple. It uses the csv library to read the document adquired from SIDC, appending certaing columns to lists according to their meaning.

## Results

The matplotlib library library was used to plot the graph. The graph has a division between northern and southern hemisphere on a parallel line to the x axys.

```python
plt.fill_between(years, north, color="#22CE00FF", alpha=0.5, label="Northern Hemisphere")
plt.fill_between(years, south, color="#CF008AFF", alpha=0.5, label="Southern Hemisphere")
ax = plt.gca()
ax.set_facecolor("#FFFFFFFF")

plt.title("Solar Cycle Hemispheric Asymmetry (1992-2026)")
plt.xlabel("Year")
plt.ylabel("Sunspot number")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
